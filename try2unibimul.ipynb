```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_5 = df.loc[df['Performance_Score']==5]
df_4 = df.loc[df['Performance_Score']==4]
df_3 = df.loc[df['Performance_Score']==3]
df_2 = df.loc[df['Performance_Score']==2]
df_1 = df.loc[df['Performance_Score']==1]
```

```python
plt.plot(df_5['Study_Hours'],np.zeros_like(df_5['Study_Hours']), 'o')
plt.plot(df_4['Study_Hours'],np.zeros_like(df_4['Study_Hours']), 'o')
plt.plot(df_3['Study_Hours'],np.zeros_like(df_3['Study_Hours']), 'o')
plt.plot(df_2['Study_Hours'],np.zeros_like(df_2['Study_Hours']), 'o')
plt.plot(df_1['Study_Hours'],np.zeros_like(df_1['Study_Hours']), 'o')
plt.xlabel('Study Hours')
plt.show()
```
