```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns   
```

```python
df = pd.read_csv(r'C:\Users\jayak\OneDrive\Desktop\Data Analytics\unibimul.csv')
```

```python
df.drop('ID', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.info
```

```python
## univariate analysis

df_A = df.loc[df['Final_Grade']=='A']
```

```python
df_A.info
```

```python
df_B = df.loc[df['Final_Grade']=='B']
df_C = df.loc[df['Final_Grade']=='C']
df_D = df.loc[df['Final_Grade']=='D']
```

```python
plt.plot(df_A['Hours_Studied'],np.zeros_like(df_A['Hours_Studied']),'o')
plt.plot(df_B['Hours_Studied'],np.zeros_like(df_B['Hours_Studied']),'o')
plt.plot(df_C['Hours_Studied'],np.zeros_like(df_C['Hours_Studied']),'o')
plt.plot(df_D['Hours_Studied'],np.zeros_like(df_D['Hours_Studied']),'o')
plt.xlabel('Hours_Studied')
plt.show()
```

```python
## Bivariate analysis

sns.FacetGrid(df,hue='Final_Grade').map(plt.scatter,'Hours_Studied','Attendance_Rate').add_legend()
```

```python
sns.FacetGrid(df,hue='Final_Grade').map(plt.scatter,'Hours_Studied','Extra_Curricular_Activities').add_legend()
```
